```python
# -*- coding: utf-8 -*-
"""KNN_Classification_Workflow.ipynb

Complete KNN Implementation - Single Code Block

# **K-Nearest Neighbors (KNN) Complete Implementation**

## **Setup and Imports**
"""

# Install and import all libraries
!pip install -q numpy pandas matplotlib seaborn scikit-learn plotly -q

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                           f1_score, roc_auc_score, confusion_matrix,
                           classification_report, roc_curve, auc,
                           precision_recall_curve, average_precision_score)
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')

# Set random seed
np.random.seed(42)
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
print("‚úÖ Libraries imported successfully!")

"""## **1. Automated Data Retrieval from GitHub**"""

# AUTOMATIC DATA DOWNLOAD FROM GITHUB
print("üì• Downloading dataset from GitHub Repository...")
GITHUB_RAW_URL = "https://raw.githubusercontent.com/Fahim-Foysal-FF/KNN-Classification-Workflow/main/data/breast_cancer.csv"


try:
    # Try GitHub first
    df = pd.read_csv(GITHUB_RAW_URL)
    print(f"‚úÖ Dataset loaded from GitHub successfully!")
except:
    try:
        # Try alternative URL
        print("üîÑ GitHub link failed, trying alternative source...")
        df = pd.read_csv(ALTERNATIVE_URL)
        print(f"‚úÖ Dataset loaded from alternative source!")
    except:
        # Last resort: Use scikit-learn built-in dataset
        print("üîÑ Using scikit-learn built-in dataset...")
        from sklearn.datasets import load_breast_cancer
        data = load_breast_cancer()
        df = pd.DataFrame(data.data, columns=data.feature_names)
        df['diagnosis'] = data.target
        # Map target to match our format
        df['diagnosis'] = df['diagnosis'].map({0: 'M', 1: 'B'})
        print("‚úÖ Dataset loaded from scikit-learn!")

print(f"üìä Dataset shape: {df.shape}")
print(f"üî§ Columns: {df.columns.tolist()[:10]}...")  # Show first 10 columns

# Save local copy
df.to_csv('breast_cancer.csv', index=False)
print("üíæ Local copy saved as 'breast_cancer.csv'")

"""## **2. Exploratory Data Analysis**"""

# EDA Section
print("\nüîç EXPLORATORY DATA ANALYSIS")
print("=" * 60)

# Display basic info
print("\nüìã Dataset Information:")
print(f"Total samples: {df.shape[0]}")
print(f"Total features: {df.shape[1] - 1}")  # Excluding target
print(f"Target column: {'diagnosis' if 'diagnosis' in df.columns else 'Diagnosis'}")

# Check for missing values
print("\n‚ùì Missing Values Analysis:")
missing_values = df.isnull().sum()
total_missing = missing_values.sum()
print(f"Total missing values: {total_missing}")
if total_missing > 0:
    print("\nColumns with missing values:")
    print(missing_values[missing_values > 0])
else:
    print("‚úÖ No missing values found!")

# Check data types
print("\nüìä Data Types:")
print(df.dtypes.value_counts())

# Display first few rows
target_col = 'diagnosis' if 'diagnosis' in df.columns else 'Diagnosis'
print(f"\nüìÑ First 5 rows (showing target column '{target_col}'):")
if target_col in df.columns:
    display(df[[target_col] + list(df.columns[:5])].head())
else:
    display(df.head())

# Target distribution visualization
plt.figure(figsize=(10, 6))
if target_col in df.columns:
    target_counts = df[target_col].value_counts()
    colors = ['#FF6B6B', '#4ECDC4']  # Red for Malignant, Teal for Benign

    # Create subplot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    # Bar plot
    bars = ax1.bar(target_counts.index, target_counts.values, color=colors, edgecolor='black', linewidth=2)
    ax1.set_title('Target Class Distribution', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Diagnosis', fontsize=12)
    ax1.set_ylabel('Count', fontsize=12)
    ax1.grid(True, alpha=0.3)

    # Add count labels on bars
    for bar in bars:
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height + 5,
                f'{int(height)}', ha='center', va='bottom', fontsize=11, fontweight='bold')

    # Pie chart
    labels = ['Malignant (M)' if str(x).upper() == 'M' else 'Benign (B)' for x in target_counts.index]
    ax2.pie(target_counts.values, labels=labels, colors=colors, autopct='%1.1f%%',
            startangle=90, explode=(0.05, 0), shadow=True)
    ax2.set_title('Class Percentage Distribution', fontsize=14, fontweight='bold')

    plt.tight_layout()
    plt.show()

    # Print statistics
    print(f"\nüéØ Target Variable Statistics:")
    print(f"Classes: {target_counts.index.tolist()}")
    print(f"Counts: {target_counts.to_dict()}")
    print(f"Percentages:")
    for cls, count in target_counts.items():
        percentage = (count / len(df)) * 100
        label = 'Malignant' if str(cls).upper() == 'M' else 'Benign'
        print(f"  {label} ({cls}): {percentage:.2f}%")

"""## **3. Data Preprocessing**"""

print("\n‚öôÔ∏è DATA PREPROCESSING")
print("=" * 60)

# Create a copy for preprocessing
df_clean = df.copy()

# Step 1: Identify target column
if 'diagnosis' in df_clean.columns:
    target_col = 'diagnosis'
elif 'Diagnosis' in df_clean.columns:
    target_col = 'Diagnosis'
else:
    # Try to find target column
    possible_targets = ['target', 'class', 'outcome', 'label']
    for col in possible_targets:
        if col in df_clean.columns:
            target_col = col
            break
    else:
        # Use last column as target
        target_col = df_clean.columns[-1]

print(f"üéØ Target column identified: '{target_col}'")

# Step 2: Handle missing values
print("\n1Ô∏è‚É£ Handling Missing Values:")
missing_before = df_clean.isnull().sum().sum()
if missing_before > 0:
    print(f"   Found {missing_before} missing values")

    # Numerical imputation
    numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) > 0:
        imputer = SimpleImputer(strategy='median')
        df_clean[numeric_cols] = imputer.fit_transform(df_clean[numeric_cols])
        print(f"   ‚úì Numerical columns imputed with median")

    # Categorical imputation
    categorical_cols = df_clean.select_dtypes(include=['object']).columns
    if len(categorical_cols) > 0:
        for col in categorical_cols:
            if col != target_col:
                df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0] if not df_clean[col].mode().empty else 'Unknown')
        print(f"   ‚úì Categorical columns imputed with mode")

    missing_after = df_clean.isnull().sum().sum()
    print(f"   Missing values after imputation: {missing_after}")
else:
    print("   ‚úì No missing values found")

# Step 3: Encode target variable
print("\n2Ô∏è‚É£ Encoding Target Variable:")
print(f"   Original values: {df_clean[target_col].unique().tolist()}")
df_clean[target_col] = df_clean[target_col].astype(str).str.upper().str.strip()

# Map to binary (Malignant=0, Benign=1)
df_clean[target_col] = df_clean[target_col].map({'M': 0, 'B': 1, 'MALIGNANT': 0, 'BENIGN': 1, '0': 0, '1': 1})
# Handle any unmapped values
df_clean[target_col] = pd.to_numeric(df_clean[target_col], errors='coerce')
df_clean = df_clean.dropna(subset=[target_col])
df_clean[target_col] = df_clean[target_col].astype(int)

print(f"   Encoded values: {df_clean[target_col].unique().tolist()}")
print(f"   Class distribution: {df_clean[target_col].value_counts().to_dict()}")

# Step 4: Separate features and target
X = df_clean.drop(columns=[target_col])
y = df_clean[target_col]

# Step 5: Encode categorical features (if any)
print("\n3Ô∏è‚É£ Encoding Categorical Features:")
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
if categorical_cols:
    print(f"   Found categorical columns: {categorical_cols}")
    X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)
    print(f"   ‚úì Applied one-hot encoding")
else:
    print("   ‚úì No categorical features found")

# Step 6: Remove constant features
print("\n4Ô∏è‚É£ Removing Constant Features:")
constant_features = [col for col in X.columns if X[col].nunique() == 1]
if constant_features:
    print(f"   Removing {len(constant_features)} constant features: {constant_features}")
    X = X.drop(columns=constant_features)
else:
    print("   ‚úì No constant features found")

# Step 7: Feature scaling (CRITICAL for KNN)
print("\n5Ô∏è‚É£ Feature Scaling:")
print("   Since KNN is distance-based, scaling is essential!")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
print(f"   ‚úì Features scaled using StandardScaler")
print(f"   Feature means after scaling: {np.mean(X_scaled, axis=0)[:3].round(2)}...")
print(f"   Feature stds after scaling: {np.std(X_scaled, axis=0)[:3].round(2)}...")

# Step 8: Train-Test Split
print("\n6Ô∏è‚É£ Train-Test Split:")
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
print(f"   Training set: {X_train.shape[0]} samples ({X_train.shape[0]/len(X)*100:.1f}%)")
print(f"   Testing set: {X_test.shape[0]} samples ({X_test.shape[0]/len(X)*100:.1f}%)")
print(f"   Training class distribution: {dict(zip(*np.unique(y_train, return_counts=True)))}")
print(f"   Testing class distribution: {dict(zip(*np.unique(y_test, return_counts=True)))}")

print("\n‚úÖ Preprocessing completed successfully!")

"""## **4. KNN Model Implementation**"""

print("\nü§ñ K-NEAREST NEIGHBORS MODEL IMPLEMENTATION")
print("=" * 60)

# Initialize KNN classifier
print("1Ô∏è‚É£ Initializing KNN Classifier:")
knn = KNeighborsClassifier(n_neighbors=5, weights='uniform', metric='minkowski', p=2)
print(f"   Initial parameters: k={knn.n_neighbors}, weights='{knn.weights}', metric='{knn.metric}'")

# Train the model
print("\n2Ô∏è‚É£ Training the model...")
knn.fit(X_train, y_train)
print("   ‚úì Model training completed")

# Make predictions
y_pred = knn.predict(X_test)
y_pred_proba = knn.predict_proba(X_test)[:, 1]

# Calculate initial metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

print("\n3Ô∏è‚É£ Initial Model Performance:")
print(f"   Accuracy:  {accuracy:.4f}")
print(f"   Precision: {precision:.4f}")
print(f"   Recall:    {recall:.4f}")
print(f"   F1-Score:  {f1:.4f}")
print(f"   ROC-AUC:   {roc_auc:.4f}")

"""## **5. Hyperparameter Tuning with GridSearchCV**"""

print("\nüéõÔ∏è HYPERPARAMETER TUNING WITH GRID SEARCH")
print("=" * 60)

# Define parameter grid
param_grid = {
    'n_neighbors': list(range(1, 31, 2)),  # Odd numbers from 1 to 29
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski'],
    'p': [1, 2]  # For minkowski metric (1=manhattan, 2=euclidean)
}

print("üîç Grid Search Parameters:")
print(f"   n_neighbors: {param_grid['n_neighbors']}")
print(f"   weights: {param_grid['weights']}")
print(f"   metric: {param_grid['metric']}")
print(f"   p: {param_grid['p']}")
print(f"   Total combinations: {len(param_grid['n_neighbors']) * len(param_grid['weights']) * len(param_grid['metric']) * len(param_grid['p'])}")

# Initialize GridSearchCV
print("\n‚è≥ Performing Grid Search (this may take a minute)...")
grid_search = GridSearchCV(
    KNeighborsClassifier(),
    param_grid,
    cv=5,  # 5-fold cross-validation
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

# Perform grid search
grid_search.fit(X_train, y_train)

print("\n‚úÖ Grid Search Completed!")
print(f"\nüéØ Best Parameters Found:")
for param, value in grid_search.best_params_.items():
    print(f"   {param}: {value}")
print(f"\nüèÜ Best Cross-validation Score: {grid_search.best_score_:.4f}")

# Get best model
best_knn = grid_search.best_estimator_

"""## **6. Elbow Method Visualization**"""

print("\nüìà ELBOW METHOD FOR OPTIMAL K")
print("=" * 60)

# Calculate error rates for different k values
error_rates = []
accuracies = []
k_values = list(range(1, 31, 2))

print("Calculating accuracy for different k values...")
for k in k_values:
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train, y_train)
    pred_temp = knn_temp.predict(X_test)
    error = 1 - accuracy_score(y_test, pred_temp)
    error_rates.append(error)
    accuracies.append(accuracy_score(y_test, pred_temp))

# Find optimal k (minimum error rate)
optimal_k = k_values[error_rates.index(min(error_rates))]
optimal_k_grid = grid_search.best_params_['n_neighbors']

print(f"   Optimal k from Elbow Method: {optimal_k}")
print(f"   Optimal k from Grid Search: {optimal_k_grid}")

# Create Elbow Method plot
plt.figure(figsize=(12, 8))
plt.plot(k_values, error_rates, marker='o', linestyle='-', color='b', linewidth=2, markersize=8, label='Error Rate')
plt.plot(k_values, accuracies, marker='s', linestyle='--', color='g', linewidth=2, markersize=8, label='Accuracy')

# Highlight optimal points
plt.axvline(x=optimal_k, color='r', linestyle=':', alpha=0.7, label=f'Elbow k={optimal_k}')
plt.axvline(x=optimal_k_grid, color='orange', linestyle=':', alpha=0.7, label=f'GridSearch k={optimal_k_grid}')

plt.xlabel('Number of Neighbors (k)', fontsize=12)
plt.ylabel('Error Rate / Accuracy', fontsize=12)
plt.title('Elbow Method: Finding Optimal k for KNN', fontsize=14, fontweight='bold')
plt.xticks(k_values)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# Additional: Accuracy vs k plot
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o', linestyle='-', color='darkgreen', linewidth=3, markersize=8)
plt.scatter(optimal_k, max(accuracies), color='red', s=200, zorder=5, label=f'Best k={optimal_k}')
plt.xlabel('Number of Neighbors (k)', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('KNN Accuracy vs Number of Neighbors', fontsize=14, fontweight='bold')
plt.xticks(k_values)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

"""## **7. Model Evaluation with Best Parameters**"""

print("\nüìä MODEL EVALUATION WITH OPTIMIZED PARAMETERS")
print("=" * 60)

# Make predictions with best model
y_pred_best = best_knn.predict(X_test)
y_pred_proba_best = best_knn.predict_proba(X_test)[:, 1]

# Calculate metrics
accuracy_best = accuracy_score(y_test, y_pred_best)
precision_best = precision_score(y_test, y_pred_best)
recall_best = recall_score(y_test, y_pred_best)
f1_best = f1_score(y_test, y_pred_best)
roc_auc_best = roc_auc_score(y_test, y_pred_proba_best)

print("üèÜ Optimized Model Performance:")
print(f"   Accuracy:  {accuracy_best:.4f}")
print(f"   Precision: {precision_best:.4f}")
print(f"   Recall:    {recall_best:.4f}")
print(f"   F1-Score:  {f1_best:.4f}")
print(f"   ROC-AUC:   {roc_auc_best:.4f}")

print("\nüìã Classification Report:")
print(classification_report(y_test, y_pred_best, target_names=['Malignant', 'Benign']))

# Performance comparison
print("\nüìà Performance Comparison:")
print("+" + "-"*45 + "+")
print("| Metric     | Before Tuning | After Tuning | Improvement |")
print("+" + "-"*45 + "+")
print(f"| Accuracy  | {accuracy:11.4f} | {accuracy_best:11.4f} | {accuracy_best-accuracy:+.4f}      |")
print(f"| Precision | {precision:11.4f} | {precision_best:11.4f} | {precision_best-precision:+.4f}      |")
print(f"| Recall    | {recall:11.4f} | {recall_best:11.4f} | {recall_best-recall:+.4f}      |")
print(f"| F1-Score  | {f1:11.4f} | {f1_best:11.4f} | {f1_best-f1:+.4f}      |")
print(f"| ROC-AUC   | {roc_auc:11.4f} | {roc_auc_best:11.4f} | {roc_auc_best-roc_auc:+.4f}      |")
print("+" + "-"*45 + "+")

"""## **8. Confusion Matrix Visualization**"""

print("\nüéØ CONFUSION MATRIX")
print("=" * 60)

# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred_best)
cm_percentage = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

# Create figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Confusion Matrix with counts
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=['Malignant (0)', 'Benign (1)'],
            yticklabels=['Malignant (0)', 'Benign (1)'])
axes[0].set_title('Confusion Matrix (Counts)', fontsize=14, fontweight='bold')
axes[0].set_ylabel('True Label', fontsize=12)
axes[0].set_xlabel('Predicted Label', fontsize=12)

# Add annotations
tn, fp, fn, tp = cm.ravel()
axes[0].text(0.5, 0.5, f'TN: {tn}', ha='center', va='center', fontsize=11, fontweight='bold', color='white')
axes[0].text(1.5, 0.5, f'FP: {fp}', ha='center', va='center', fontsize=11, fontweight='bold')
axes[0].text(0.5, 1.5, f'FN: {fn}', ha='center', va='center', fontsize=11, fontweight='bold')
axes[0].text(1.5, 1.5, f'TP: {tp}', ha='center', va='center', fontsize=11, fontweight='bold', color='white')

# Plot 2: Confusion Matrix with percentages
sns.heatmap(cm_percentage, annot=True, fmt='.2%', cmap='Greens', ax=axes[1],
            xticklabels=['Malignant (0)', 'Benign (1)'],
            yticklabels=['Malignant (0)', 'Benign (1)'])
axes[1].set_title('Confusion Matrix (Percentages)', fontsize=14, fontweight='bold')
axes[1].set_ylabel('True Label', fontsize=12)
axes[1].set_xlabel('Predicted Label', fontsize=12)

plt.tight_layout()
plt.show()

# Print confusion matrix details
print("\nüìä Confusion Matrix Details:")
print(f"True Negatives (TN): {tn} - Correctly predicted Malignant")
print(f"False Positives (FP): {fp} - Malignant incorrectly predicted as Benign")
print(f"False Negatives (FN): {fn} - Benign incorrectly predicted as Malignant")
print(f"True Positives (TP): {tp} - Correctly predicted Benign")

# Calculate additional metrics
specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
npv = tn / (tn + fn) if (tn + fn) > 0 else 0  # Negative Predictive Value
print(f"\nüìà Additional Metrics:")
print(f"Specificity (True Negative Rate): {specificity:.4f}")
print(f"Negative Predictive Value (NPV): {npv:.4f}")

"""## **9. ROC Curve and AUC**"""

print("\nüìà ROC CURVE AND AUC")
print("=" * 60)

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba_best)
roc_auc = auc(fpr, tpr)

# Find optimal threshold (Youden's J statistic)
youden_j = tpr - fpr
optimal_idx = np.argmax(youden_j)
optimal_threshold = thresholds[optimal_idx]

print(f"ROC-AUC Score: {roc_auc:.4f}")
print(f"Optimal Threshold: {optimal_threshold:.4f}")

# Create ROC curve plot
plt.figure(figsize=(12, 8))

# Plot ROC curve
plt.plot(fpr, tpr, color='darkorange', lw=3, label=f'ROC curve (AUC = {roc_auc:.4f})')

# Plot diagonal line (random classifier)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')

# Highlight optimal point
plt.scatter(fpr[optimal_idx], tpr[optimal_idx], color='red', s=200,
           zorder=5, label=f'Optimal Threshold ({optimal_threshold:.3f})')

# Add text for optimal point
plt.text(fpr[optimal_idx] + 0.02, tpr[optimal_idx] - 0.02,
        f'TPR={tpr[optimal_idx]:.3f}, FPR={fpr[optimal_idx]:.3f}',
        fontsize=11, bbox=dict(facecolor='white', alpha=0.8))

# Customize plot
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=12)
plt.ylabel('True Positive Rate (Sensitivity/Recall)', fontsize=12)
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=14, fontweight='bold')
plt.legend(loc="lower right")
plt.grid(True, alpha=0.3)

# Add AUC score in the plot
plt.text(0.6, 0.2, f'AUC = {roc_auc:.4f}\nOptimal Threshold = {optimal_threshold:.3f}',
        fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()

# Precision-Recall Curve
precision_vals, recall_vals, _ = precision_recall_curve(y_test, y_pred_proba_best)
avg_precision = average_precision_score(y_test, y_pred_proba_best)

plt.figure(figsize=(12, 6))
plt.plot(recall_vals, precision_vals, color='green', lw=3, label=f'PR curve (AP = {avg_precision:.4f})')
plt.xlabel('Recall', fontsize=12)
plt.ylabel('Precision', fontsize=12)
plt.title('Precision-Recall Curve', fontsize=14, fontweight='bold')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.grid(True, alpha=0.3)
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()

"""## **10. 2D Decision Boundary Visualization**"""

print("\nüé® 2D DECISION BOUNDARY VISUALIZATION")
print("=" * 60)

print("Reducing dimensions to 2D using PCA for visualization...")

# Reduce to 2D using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Split PCA data
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(
    X_pca, y, test_size=0.2, random_state=42, stratify=y
)

# Train KNN on 2D data
knn_2d = KNeighborsClassifier(
    n_neighbors=best_knn.n_neighbors,
    weights=best_knn.weights,
    metric=best_knn.metric,
    p=best_knn.p if hasattr(best_knn, 'p') else 2
)
knn_2d.fit(X_train_pca, y_train_pca)

# Create mesh grid
h = 0.02  # step size in mesh
x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict for each point in mesh
Z = knn_2d.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Create visualization
plt.figure(figsize=(14, 10))

# Plot decision boundary
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.colorbar(label='Decision Boundary')

# Plot training points
scatter_train = plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1],
                           c=y_train_pca, cmap='coolwarm',
                           edgecolors='k', s=100, alpha=0.7,
                           label='Training Points')

# Plot test points
scatter_test = plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1],
                          c=y_test_pca, cmap='coolwarm',
                          marker='s', edgecolors='k', s=150,
                          alpha=1.0, linewidth=2, label='Test Points')

# Add legend
legend1 = plt.legend(*scatter_train.legend_elements(),
                     title="Classes", loc='upper right')
plt.gca().add_artist(legend1)
plt.legend(loc='lower right')

# Customize plot
plt.xlabel(f'Principal Component 1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)', fontsize=12)
plt.ylabel(f'Principal Component 2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)', fontsize=12)
plt.title(f'KNN Decision Boundary (k={knn_2d.n_neighbors}, {knn_2d.weights} weights)',
          fontsize=14, fontweight='bold')

# Add accuracy on 2D space
y_pred_2d = knn_2d.predict(X_test_pca)
accuracy_2d = accuracy_score(y_test_pca, y_pred_2d)
plt.text(0.02, 0.98, f'2D Accuracy: {accuracy_2d:.4f}',
         transform=plt.gca().transAxes, fontsize=12,
         bbox=dict(facecolor='white', alpha=0.8))

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"PCA Explained Variance: PC1={pca.explained_variance_ratio_[0]*100:.1f}%, PC2={pca.explained_variance_ratio_[1]*100:.1f}%")
print(f"Total Variance Explained: {(pca.explained_variance_ratio_[0] + pca.explained_variance_ratio_[1])*100:.1f}%")
print(f"2D KNN Accuracy: {accuracy_2d:.4f}")

"""## **11. Feature Importance Analysis**"""

print("\nüîç FEATURE IMPORTANCE ANALYSIS")
print("=" * 60)

# Since KNN doesn't have built-in feature importance, we use permutation importance
from sklearn.inspection import permutation_importance

print("Calculating permutation importance...")
perm_importance = permutation_importance(
    best_knn, X_test, y_test,
    n_repeats=10,
    random_state=42,
    n_jobs=-1
)

# Create dataframe with importance scores
feature_importance_df = pd.DataFrame({
    'feature': X.columns,
    'importance_mean': perm_importance.importances_mean,
    'importance_std': perm_importance.importances_std
}).sort_values('importance_mean', ascending=False)

print("\nüìä Top 10 Most Important Features:")
print(feature_importance_df.head(10).to_string(index=False))

# Visualize feature importance
plt.figure(figsize=(12, 8))
top_features = feature_importance_df.head(15)
plt.barh(range(len(top_features)), top_features['importance_mean'],
         xerr=top_features['importance_std'],
         color='steelblue', ecolor='black', capsize=5)
plt.yticks(range(len(top_features)), top_features['feature'])
plt.xlabel('Permutation Importance', fontsize=12)
plt.title('Top 15 Features by Permutation Importance', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()

"""## **12. Cross-Validation Results**"""

print("\nüìä CROSS-VALIDATION RESULTS")
print("=" * 60)

# Perform 10-fold cross-validation
cv_scores = cross_val_score(best_knn, X_scaled, y, cv=10, scoring='accuracy')

print("10-Fold Cross-Validation Scores:")
for i, score in enumerate(cv_scores, 1):
    print(f"  Fold {i}: {score:.4f}")

print(f"\nüìà Cross-Validation Statistics:")
print(f"  Mean Accuracy: {cv_scores.mean():.4f}")
print(f"  Std Deviation: {cv_scores.std():.4f}")
print(f"  Min Accuracy:  {cv_scores.min():.4f}")
print(f"  Max Accuracy:  {cv_scores.max():.4f}")
print(f"  95% CI: [{cv_scores.mean() - 1.96*cv_scores.std():.4f}, {cv_scores.mean() + 1.96*cv_scores.std():.4f}]")

# Visualize CV results
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), cv_scores, marker='o', linestyle='-', color='blue', linewidth=2, markersize=8)
plt.axhline(y=cv_scores.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {cv_scores.mean():.4f}')
plt.fill_between(range(1, 11), cv_scores.mean() - cv_scores.std(),
                  cv_scores.mean() + cv_scores.std(), alpha=0.2, color='gray', label='¬±1 Std Dev')
plt.xlabel('Fold Number', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('10-Fold Cross-Validation Results', fontsize=14, fontweight='bold')
plt.xticks(range(1, 11))
plt.ylim([0.85, 1.0])
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

"""## **13. Sample Predictions with Probabilities**"""

print("\nüîÆ SAMPLE PREDICTIONS WITH PROBABILITIES")
print("=" * 60)

# Select random samples from test set
np.random.seed(42)
sample_indices = np.random.choice(len(X_test), min(15, len(X_test)), replace=False)

# Create results dataframe
results_df = pd.DataFrame({
    'Sample_ID': sample_indices,
    'True_Label': y_test.iloc[sample_indices].values,
    'Predicted_Label': y_pred_best[sample_indices],
    'Probability_Malignant': best_knn.predict_proba(X_test.iloc[sample_indices])[:, 0],
    'Probability_Benign': y_pred_proba_best[sample_indices],
    'Correct': y_test.iloc[sample_indices].values == y_pred_best[sample_indices]
})

# Map labels to readable names
results_df['True_Label_Name'] = results_df['True_Label'].map({0: 'Malignant', 1: 'Benign'})
results_df['Predicted_Label_Name'] = results_df['Predicted_Label'].map({0: 'Malignant', 1: 'Benign'})

print("Sample Predictions (15 random test samples):")
print("+" + "-"*95 + "+")
print("| Sample | True      | Predicted | Prob(Malignant) | Prob(Benign) | Correct | Confidence |")
print("+" + "-"*95 + "+")

for _, row in results_df.iterrows():
    correct_symbol = "‚úì" if row['Correct'] else "‚úó"
    confidence = max(row['Probability_Malignant'], row['Probability_Benign'])
    confidence_level = "HIGH" if confidence > 0.8 else "MED" if confidence > 0.6 else "LOW"

    print(f"| {row['Sample_ID']:6d} | {row['True_Label_Name']:9s} | {row['Predicted_Label_Name']:9s} | "
          f"{row['Probability_Malignant']:.4f}       | {row['Probability_Benign']:.4f}    | "
          f"{correct_symbol:^7s} | {confidence_level:^10s} |")

print("+" + "-"*95 + "+")

# Calculate statistics
print(f"\nüìä Prediction Statistics:")
print(f"  Total samples shown: {len(results_df)}")
print(f"  Correct predictions: {results_df['Correct'].sum()} ({(results_df['Correct'].sum()/len(results_df)*100):.1f}%)")
print(f"  Incorrect predictions: {(~results_df['Correct']).sum()} ({((~results_df['Correct']).sum()/len(results_df)*100):.1f}%)")
print(f"  Average confidence: {results_df[['Probability_Malignant', 'Probability_Benign']].max(axis=1).mean():.4f}")

"""## **14. Final Summary and Model Saving**"""

print("\n" + "="*70)
print("üéâ KNN IMPLEMENTATION COMPLETE - FINAL SUMMARY")
print("="*70)

# Create summary table
summary_data = {
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC', 'Optimal k', 'Best Weights', 'Best Metric'],
    'Value': [
        f"{accuracy_best:.4f}",
        f"{precision_best:.4f}",
        f"{recall_best:.4f}",
        f"{f1_best:.4f}",
        f"{roc_auc_best:.4f}",
        f"{best_knn.n_neighbors}",
        f"{best_knn.weights}",
        f"{best_knn.metric}"
    ],
    'Interpretation': [
        f"{accuracy_best*100:.1f}% of predictions are correct",
        f"{precision_best*100:.1f}% of predicted Benign are actually Benign",
        f"{recall_best*100:.1f}% of actual Benign are correctly identified",
        "Harmonic mean of precision and recall",
        "Excellent discriminative ability" if roc_auc_best > 0.9 else "Good discriminative ability",
        f"Using {best_knn.n_neighbors} nearest neighbors",
        f"Weighting: {best_knn.weights}",
        f"Distance metric: {best_knn.metric}"
    ]
}

summary_df = pd.DataFrame(summary_data)
print("\nüìã FINAL MODEL PERFORMANCE SUMMARY:")
print(summary_df.to_string(index=False))

# Save the model
import joblib
import pickle

print("\nüíæ SAVING MODEL AND PREPROCESSORS...")

# Save model
joblib.dump(best_knn, 'best_knn_model.pkl')
print("  ‚úì KNN model saved as 'best_knn_model.pkl'")

# Save scaler
joblib.dump(scaler, 'scaler.pkl')
print("  ‚úì Scaler saved as 'scaler.pkl'")

# Save PCA transformer
joblib.dump(pca, 'pca_transformer.pkl')
print("  ‚úì PCA transformer saved as 'pca_transformer.pkl'")

# Save results
results = {
    'accuracy': accuracy_best,
    'precision': precision_best,
    'recall': recall_best,
    'f1': f1_best,
    'roc_auc': roc_auc_best,
    'best_params': grid_search.best_params_,
    'feature_importance': feature_importance_df.to_dict()
}

with open('knn_results.pkl', 'wb') as f:
    pickle.dump(results, f)
print("  ‚úì Results saved as 'knn_results.pkl'")

print("\n" + "="*70)
print("‚úÖ PROJECT COMPLETED SUCCESSFULLY!")
print("="*70)
print("\nüìÅ Files created:")
print("  - best_knn_model.pkl (Trained KNN model)")
print("  - scaler.pkl (Fitted StandardScaler)")
print("  - pca_transformer.pkl (Fitted PCA)")
print("  - knn_results.pkl (All results and metrics)")
print("  - breast_cancer.csv (Original dataset)")

print("\nüéØ Key Insights:")
print(f"  1. Optimal k-value: {best_knn.n_neighbors}")
print(f"  2. Best distance metric: {best_knn.metric}")
print(f"  3. Model accuracy: {accuracy_best*100:.2f}%")
print(f"  4. ROC-AUC Score: {roc_auc_best:.4f} (Excellent)")

print("\nüöÄ The model is ready for deployment!")
```
